# Entrenar modelo
<div align="center"><img src="https://raw.githubusercontent.com/InoveAlumnos/dataset_analytics_python/master/images/Pipeline4.png" width="1000" align="middle"></div>

```python
# Almuno: Crear un numpy array "X" con los features de entrada excepto el nombre del pais (country)

X = df.drop('country', axis=1).values #en el vector X, todas las columnas -features- MENOS el nombre del país
```

#### Crear un modelo de segmentación con Kmeans
Parámetros
- n_clusters --> (k) número de clusters/grupos (defecto 5)
- init --> método utilizado para determianr donde comienzan los clusters
 - k-means++ --> mecanismo inteligente para determinar el comienzo (defecto)
 - random --> los centros se determinarn aleatoriamente
- max_iter --> cantidad de iteración (defecto 300)

```python
# Almuno: Entrenar un modelo de Kmeans en un rango de cluster (2, 10)
# Conservar el resultado de "inertial" para utilizar como métrica de seleccion

#--> Este procedimiento es utilizado para encontrar la cantidad óptima de clusters, entrenando distintos modelos de Kmeans (variando sus parámetros n_cluster sucesivamente,
# en cada iteración

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

nro_clusters = list(range(2,11))

# Creamos 2 vectores nulos del tamaño de len(nro_clusters) que se irán sobreescribiendo: uno, con los errores; otro, con el score, para cada una de las 9 pruebas. 
vec_error = np.zeros((len(nro_clusters)))
vec_score = np.zeros((len(nro_clusters)))

for i in range(len(nro_clusters)):
    #para cada i, se crea un objeto de la clase KMeans. El parámetro que varía en cada iteración, será el nro de clusters según la lista definida con ese mismo nombre
    kmeans = KMeans(n_clusters=nro_clusters[i], init="k-means++", random_state=0)
    #se entrena el objeto creado -el modelo de segmentación-
    kmeans.fit(X)

    #se guardan errores y scores en la posición correspondiente para cada vector, a medida que se ejecuta cada iteración.
    vec_error[i] = kmeans.inertia_
    vec_score[i]= silhouette_score(X, kmeans.labels_)

```

```python
# Dibujar la evolucion del error (inertia) y deteminar el 
# punto de quiebre (elbow point)

fig = plt.figure()
ax = fig.add_subplot()

# Graficamos ambas curvas en un mismo sistema de ejes cartesianos: error y score
# n_clusters = [2, 3, 4, 5, 6, 7, 8, 9, 10]
# vec_error/vec_error.max(), divide cada error por el máximo error
ax.plot(nro_clusters, vec_error/vec_error.max(), label='error');
ax.plot(nro_clusters, vec_score, label ='score')
ax.grid('dashed', color = 'lightgray')
plt.xlabel('Nro. de clusters', color='darkblue')
plt.ylabel('Resultados', color='darkblue')

ax.legend()
plt.show()

best_cluster = nro_clusters[vec_score.argmax()]
print(f'\nEl resultado más óptimo se consigue con {best_cluster} clusters.\nEl mismo, se deduce del elbow point de la curva error.')
```

Se puede observar que el punto de elbow podría ser con cluster=4, pero ya sabemos de ante mano que queremos dividir el dataset en 3 grupos, utilizaremos n_clusters=3

```python
# Alumno: Entrenar el modelo con el mejor n_clusters=3
# y determinar los labels

# Como sabemos que queremos utilizar k=3 como nro de clusters, sobreescribo el objeto kmeans (la clase recibe el k deseado como parámetro)
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)

# lo entreno
kmeans.fit(X)

# determino los labels
labels = kmeans.predict(X)
```

```python
# Alumno: Sumar la columna de labels con el nombre de countryseg (customer segmentation)
# a su dataset

# Si bien antes no lo necesité -no fueron modificados los datos del dataset original-, ahora usaré una copia del mismo, para adicionar la columna pedida
df2 =df.copy()

df2['CountrySeg'] = labels
df2
```
